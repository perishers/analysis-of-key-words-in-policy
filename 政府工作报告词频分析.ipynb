```python
frequency_lists = []
key_words = '化解、防范化解、遏制、守住、管控 、风险、重大风险、政府债务风险、金融风险、金融稳定、风险隐患、隐性债务、专项债券、系统性区域性金融风险的底线、金融监管、债务总体可控'
key_words_list = key_words.split('、')
for i in range(2008,2024):
    # 打开一个.txt文件以供读取
    with open(r'C:\Users\15812\Desktop\哲社科\哲社科\{year}.txt'.format(year=i), 'r', encoding='utf-8') as file:
        content = file.read()
        frequency_list = []
        frequency_list.append('{year}政府工作报告'.format(year=i))
        
        for word in key_words_list:
            word_frequency = content.count(word)
            frequency_list.append(word_frequency)
        frequency_lists.append(frequency_list)
print(frequency_lists)
```

```python
import pandas as pd

df = pd.DataFrame(frequency_lists)

df.columns = ['年份','化解', '防范化解', '遏制', '守住', '管控 ', '风险', '重大风险', '政府债务风险', '金融风险', '金融稳定', '风险隐患', '隐性债务', '专项债券', '系统性区域性金融风险的底线', '金融监管', '债务总体可控']
df
```
